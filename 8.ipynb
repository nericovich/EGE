```python
from itertools import *
cnt =0 
for i in product('весна', repeat=5):
    s=''.join(i)
    if s[0] == 'н':
        if s.count('в')== 2:
            if s.count('е') <= 1 and s.count('с') <= 1 and s.count('н') <= 1 and s.count('а') <= 1:
                cnt +=1 

cnt 
```

```python
s = '0123456789awe'
cnt =0
for i in product(s, repeat=6):
    t= ''.join(i)
    if t[0] != '0':
        if t.count('5') <= 1:
            t = t.replace('1','q').replace('3','q').replace('5','q').replace('7','q').replace('9','q').replace('w','q')
            if not 'qq' in t:
                cnt +=1

cnt
```
