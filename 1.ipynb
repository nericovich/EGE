```python
from itertools import *
graph = 'AE AF EB BF FG GD GH HC CB'
tabl = '10 12 15 5 7 20 8 13 14'

gr = set(graph.split())

for  per in permutations('ABCDEFGH'):
    new = tabl.replace('1', per[0]).replace('2', per[1]).replace('3', per[2]).replace('4', per[3]).replace('5', per[4]).replace('6', per[5]).replace('7', per[6])
    if gr == set(new.split()):
        print('1 2 3 4 5 6 7')
        print(*per)
```
