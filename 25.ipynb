```python
from  fnmatch import *

def f(n):
    s = 0
    while n > 0:
        s += (n% 10)
        n //= 10
    return s
cnt = 0
for i in range(17999902399, 1, -1):
    if fnmatch(str(i), '17*023?9'):
        if f(i) % 11 == 0:
            print(i, f(i)//11)
            cnt +=1
        if cnt == 5:
            break




```

```python
def f(n):
    for i in range(2, int(n**0.5) + 1):
        if n % i ==0:
            return 0
    return 1
bo = []
m, n = map(int, input().split())
num1 = []
for i in range(m, n+1):
    tec = []
    if f(i):
        a = i
        while i > 0:
            num1.append(i%10)
            i //=10
        num1.sort()
        for j in range(i, n+1):
            if f(j):
                b = j
                num2 = []
                while j>0:
                    num2.append(j% 10)
                    j //= 10
                num2.sort()
        if (num1 == num2) and (a not in bo) and (b not in bo):
            tec.append(a)
            tec.append(b)
            bo.append(a)
            bo.append(b)
    if len(tec) > 1:
        print(tec)
```
