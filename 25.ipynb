```python
from  fnmatch import *

def f(n):
    s = 0
    while n > 0:
        s += (n% 10)
        n //= 10
    return s
cnt = 0
for i in range(17999902399, 1, -1):
    if fnmatch(str(i), '17*023?9'):
        if f(i) % 11 == 0:
            print(i, f(i)//11)
            cnt +=1
        if cnt == 5:
            break




```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


m, n = map(int, input().split())

groups = {}

for x in range(m, n + 1):
    if is_prime(x):
        key = tuple(sorted(str(x)))
        groups.setdefault(key, []).append(x)

for group in groups.values():
    if len(group) > 1:
        print(group)
```
