```python
from  fnmatch import *

def f(n):
    s = 0
    while n > 0:
        s += (n% 10)
        n //= 10
    return s
cnt = 0
for i in range(10**10, 1, -1):
    if fnmatch(str(i), '17*023?9'):
        if f(i) % 11 == 0:
            print(i, f(i)//11)
            cnt +=1
        if cnt == 5:
            break




```
