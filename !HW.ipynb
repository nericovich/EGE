```python
def dfs(v):
    visited[v] = True
    for u in range(n):
        if matrix[v][u] == 1 and not visited[u]:
            dfs(u)

n = int(input("Введите число вершин: "))

matrix = []
print("Введите матрицу смежности:")
for _ in range(n):
    matrix.append(list(map(int, input().split())))

visited = [False] * n
components = 0

for i in range(n):
    if not visited[i]:
        dfs(i)
        components += 1

print("Число компонент связности:", components)
```
