```python
with open('17_1.txt') as f:
    num = [int(x) for x in f.readlines()]
mini = 100001
for i in num:
    if (i % 25 == 0) and i < mini:
        mini = i
sums = []
cnt = 0
for i in range(len(num) -1):
    if num[i] % 2 ==0 or num[i+1] % 2 ==0:
        if (num[i] + num[i+1]) % mini==0:
            cnt +=1
            sums.append(num[i] + num[i+1])

max(sums)
```

```python
with open('17_2.txt') as f:
    num = [int(x) for x in f.readlines()]
mini = 100001
for i in num:
    if (i % 45 == 0) and i < mini and len(str(i))==3: 
        mini = i

sums = []
cnt = 0
for i in range(len(num) - 1):
    if len(str(num[i])) == 4 or len(str(num[i+1])) == 4:
        if (num[i] + num[i+1]) % mini==0:
            cnt +=1
            sums.append(num[i] + num[i+1])

min(sums)
```
