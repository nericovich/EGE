```python
def d(x, y):
    return (x **2 + y ** 2) ** 0.5

def f(x, y, p):
    if d(x, y)  > 14 and (p == 4 or p == 2): 
        return True
    if d(x, y)  > 14 and p < 4:
        return False
    if d(x, y)  <= 14 and p == 4:
        return False
    
    h = [f(2*x,y,p+1), f(x, y+3, p+1), f(x, y+4, p+1)]

    if p % 2 == 1:
        return any(h)
    else:
        return all(h)


def f1(x, y, p):
    if d(x, y)  > 14 and p == 2: 
        return True
    if d(x, y)  > 14 and p < 2:
        return False
    if d(x, y)  <= 14 and p == 2:
        return False
    
    h = [f1(2*x,y,p+1), f1(x, y+3, p+1), f1(x, y+4, p+1)]

    if p % 2 == 1:
        return any(h)
    else:
        return all(h)
    


for s in range(1, 14):
    if f(3, s, 0) and not f1(3, s, 0):
        print(s)
```
