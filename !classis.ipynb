```python
from fnmatch import fnmatch
def s(n):
    r= 0
    while n >0:
        r+= (n % 10)
        n//=10
    return r
cnt = 0
for i in range(700011, 2 *10**8):
    if not fnmatch(str(i), '*0?3*') and not fnmatch(str(i), '*4?2') and not fnmatch(str(i), '*1*') and i % 13 ==0:
        print(i, s(i))
        cnt +=1
    if cnt == 5:
        break
```
