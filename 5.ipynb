```python
cnt = 0

for i in range(300, 401):
    num = []
    while i > 0:
        num.append(i% 10)
        i //=10
    num = sorted(num, reverse=True)
    maxi = str(num[0]) + str(num[1])
    if num[2] != 0:
        mini = str(num[2]) + str(num[1])
    else:
        mini = str(num[1]) + str(num[0])
    if int(maxi)-int(mini) == 20:
        cnt +=1

cnt
```
