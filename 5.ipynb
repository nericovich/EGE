```python
cnt =0
for n in range(1000, 10000):
    num =[]
    s = ''
    while n >0:
        num.append(n% 10)
        n //=10
    
    num.sort()
    s = str(num[0] + num[3]) + str(num[1]*num[2])
    if int(s) == 1128:
        cnt += 1

cnt
    
```

```python
cnt = 0

for i in range(300, 401):
    num = []
    while i > 0:
        num.append(i% 10)
        i //=10
    num = sorted(num, reverse=True)
    maxi = str(num[0]) + str(num[1])
    if num[2] != 0:
        mini = str(num[2]) + str(num[1])
    else:
        mini = str(num[1]) + str(num[0])
    if int(maxi)-int(mini) == 20:
        cnt +=1

cnt
```

```python
def f(n):
    s = ''
    while n >0:
        s = str(n % 8) + s
        n //=8
    return s

def g(n):
    c = 0
    while n >0:
        c += n % 10
        n //=10
    return c

r = []

for i in range(483, 1000):
    m = f(i)
    if g(int(m)) % 2 ==0:
        r.append(int(m + f(g(int(m))), 8))
    else:
        r.append(int(f(g(int(m))) + m, 8))

r.sort()
r
```
