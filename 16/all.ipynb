```python
from functools import lru_cache
@lru_cache(None)
def f(n):
    if n >10:
        return 2 **n +3
    else:
        return f(n+2) + 2*f(n+1)

cnt =0
for i in range(-1000, 1001):
    if f(i) % 3 == 0:
        cnt += 1

cnt
```
