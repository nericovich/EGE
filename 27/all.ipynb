```python
with open('27_А_1.txt') as f:

    a = [(x) for x in f.readlines()]
x1 = []
y1 = []
x2= []
y2 =[]
for i in range(len(a)):
    a[i] = a[i].split()

for i in range(len(a)):
    if float(a[i][1]) < 3:
        x1.append(float(a[i][0]))
        y1.append(float(a[i][1]))
    else:
        x2.append(float(a[i][0]))
        y2.append(float(a[i][1]))
```

```python

su = 90999999999999999999
a1= 0
dix = []
diy = []
ds = 0
for ch in range(len(x1)):
    s = 0
    for i in range(len(x1)):
        d = ((x1[ch] - x1[i]) ** 2 + (y1[ch] - y1[i]) ** 2) ** 0.5
        s += d
    if su > s:
        su = s 
        a1 = ch
print(x1[a1], y1[a1])
print(max(dix)* 10000)
print(max(diy)* 10000)
```
