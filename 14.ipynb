```python
def f(n):
    s = ''
    while n > 0:
        s = str(n % 5) + s
        n //= 5
    return s
maxi = 0

for x in range(1, 2031):
    num = 5 ** 100 - x
    if f(num).count('0') > maxi:
        maxi = f(num).count('0')
        c = x
        print(c)
x
```
